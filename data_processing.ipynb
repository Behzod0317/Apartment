#  Format the date  (year, month)

```python
df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
df['publication_date'] = df['publication_date'].dt.strftime('%Y, %B')
```

```python
df
```

# Edit the texts to all lowercase in the ‘text’ column

```python
df['text'] = df['text'].str.lower()
```

```python
df
```

# rubric + subrubric

```python
df['new_rubric'] = df['rubric'] + ', ' + df['subrubric']
df['new_rubric'] = df['new_rubric'].astype(str)
```
